```python
# Drop rows with any null values
base_dataset = base_dataset.dropna()
print(base_dataset.shape)
# Verify if there are any null values left
print(f"are there anymore nulls? {base_dataset.isnull().sum().any()}")  # Should print 0 for all columns
```

```python
# convert string cols to categorical for future encoding
categorical_cols = [base_dataset.columns[i] for i, x in enumerate(base_dataset.dtypes) if x == 'object']
base_dataset[categorical_cols] = base_dataset[categorical_cols].astype('category')
# Convert categorical to label encodings
for col in categorical_cols:
    base_dataset[col] = base_dataset[col].cat.codes
base_dataset.to_csv("../data/baseline_model_dataset.csv",index=False)
base_dataset.head()
```
