# Prepare Dataset (minimal preprocessing)

- apply label encoding

Surprisingly the columwise data distributions remained the same, unclear/(didn't check) if the feature relationships are still in tact

```python
# Drop rows with any null values
base_dataset = base_dataset.dropna()
print(base_dataset.shape)
# Verify if there are any null values left
print(f"are there anymore nulls? {base_dataset.isnull().sum().any()}")  # Should print 0 for all columns
```

```python
# convert string cols to categorical for future encoding
categorical_cols = [base_dataset.columns[i] for i, x in enumerate(base_dataset.dtypes) if x == 'object']
base_dataset[categorical_cols] = base_dataset[categorical_cols].astype('category')

#make copies for the encodings
base_dataset_label_encoded = base_dataset.copy(deep=True)
base_dataset_one_hot = base_dataset.copy(deep=True)
```

```python
# Convert categorical to label encodings
for col in categorical_cols:
    base_dataset_label_encoded[col] = base_dataset_label_encoded[col].cat.codes
base_dataset_label_encoded.to_csv("../data/baseline_dataset.csv",index=False)
base_dataset_label_encoded.head()
```

```python
## convert categorical to one hot encodings
base_dataset_one_hot = pd.get_dummies(base_dataset_one_hot, drop_first=True)
base_dataset_one_hot.to_csv("../data/baseline_dataset_one_hot.csv",index=False)
base_dataset_one_hot.head()
```
